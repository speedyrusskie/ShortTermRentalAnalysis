```python
#import appropriate libraries to clean up dataset
import pandas as pd
```

```python
#create dataframe
scraped_listings = pd.read_csv(r'C:\Users\Daniel\Data Science Projects\20190822 Withheld\1. Original Data\scraped_listings - Copy.csv', sep =',',error_bad_lines = False, encoding = 'latin-1')
```

```python
#get an idea of how data looks like and what columns to drop
scraped_listings.head(10)
```

```python
#see the number of rows and columns in data set
scraped_listings.shape
```

```python
#see the names of each individual column
scraped_listings.columns
```

```python
#count the number of filled-in rows. Looks like all rows have data in them!
scraped_listings.count()
```

```python
#drop listing column
scraped_listings = scraped_listings.drop('listing', axis = 1)
```

```python
#drop mapped_location column
scraped_listings = scraped_listings.drop('mapped_location', axis = 1)
```

```python
#drop name column
scraped_listings = scraped_listings.drop('name', axis = 1)
```

```python
#drop host name column
scraped_listings = scraped_listings.drop('host_name', axis = 1)
```

```python
#see resulting dataset. it is much more workable now
scraped_listings.head(10)
```

```python
#save to csv file
scraped_listings.to_csv(r'C:\Users\Daniel\Data Science Projects\20190822 Withheld\2. Prepared Data\scraped_listings_checked.csv', index = None, header =True)
```
